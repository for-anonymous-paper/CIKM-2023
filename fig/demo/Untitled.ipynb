```python
from numpy import ma
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import ticker, cm

%matplotlib inline
x, y = np.meshgrid(np.arange(0, 10.1, 0.1),
                         np.arange(0, 10.1, 0.1))

h = x * y
h = np.minimum(h, 30.26)
z = ma.masked_where(h <= 30.25, h)

fig, ax = plt.subplots()
cs = ax.contourf(x, y, z, colors = 'lightblue', linewidth = 2, linestyles = 'dashed', extend='both')
linex = np.arange(1.1, 10.1, 0.1)
liney = -linex + 11
ax.plot(linex, liney, linestyle = 'dashed', color = 'black')


lx = np.arange(3.1, 10, 0.1)
ly = 30.26 / lx
ax.plot(lx, ly, color = 'b', linewidth = 2)

ax.text(7, 7, "Decision \n space", fontsize = 15)
ax.annotate('Skyline', xy=(8.6,3.5), xytext=(5, 2), fontsize = 15,
            arrowprops=dict(facecolor='black', shrink=0.05))
ax.plot(5.5, 5.5, color = 'red', marker = "o", markersize = 12, markerfacecolor = "none")
ax.plot(1.5, 1.5, color = 'black', marker = "o", markersize = 5)#, markerfacecolor = "none")
ax.text(1, 2, "Input", fontsize = 15)
ax.set_xlabel('Salary', fontsize = 15)
ax.set_ylabel('Education year', fontsize = 15)
ax.set_xticks(())
ax.set_yticks(())

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.plot(1, 0, ">k", transform=ax.get_yaxis_transform(), clip_on=False)
ax.plot(0, 1, "^k", transform=ax.get_xaxis_transform(), clip_on=False)

plt.savefig("MMO_Demo.eps")
plt.show()
```
